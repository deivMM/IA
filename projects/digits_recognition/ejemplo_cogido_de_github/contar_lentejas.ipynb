```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def frame_imshow_vis(img, tit = 'image'):
    cv2.imshow(tit, img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()

def frame_plt_vis(img, fsize = (8, 8)):
    f, ax = plt.subplots(figsize=fsize)
    ax.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    ax.axis('off')
    return ax
```

```python
image = cv2.imread('images/lentejas.jpeg')
frame_plt_vis(image);
```

```python
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
blur = cv2.GaussianBlur(gray, (11, 11), 0)
# _, umbral = cv2.threshold(blur, 120, 255, cv2.THRESH_BINARY)

canny = cv2.Canny(blur, 80, 160)

frame_plt_vis(canny);

```
