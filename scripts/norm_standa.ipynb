```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
```

### Normalization
Also know as Min-Max scaling. Features are rescaled between 0 and 1.

$$X_{{norm}} = \frac{X - X_{{min}}}{X_{{max}} - X_{{min}}}$$

### Standardization
Also known as z-score normalization.
$$X_{standardization}=\frac{X-\mu}{\sigma}$$



```python
data = pd.read_csv('../data/garay_data.csv')

data_pot = data['ZOE5__potencia'][data['ZOE5__potencia'].between(50,200)]

norm_scaler = MinMaxScaler()
data_nom = norm_scaler.fit_transform(data_pot.values.reshape(-1, 1))

std_scaler = StandardScaler()
data_std = std_scaler.fit_transform(data_pot.values.reshape(-1, 1))

plt.hist(data_pot, bins=50)
plt.show()

plt.hist(data_nom, bins=50)
plt.hist(data_std, bins=50, zorder=-1)
plt.show()
```
